```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

#Masking Manually Wall

```python
image = cv2.imread('wall4.jpeg')
```

```python
# Define the color threshold (in BGR format)
lower_bound = np.array([20, 20, 20])
upper_bound = np.array([120, 120, 120])

mask = cv2.inRange(image, lower_bound, upper_bound)
```

```python
# Create a copy of the original image to highlight the masked areas
highlighted_image = image.copy()
highlighted_image[mask == 255] = [185, 174, 170]  # Highlight the pixels with a chosen color
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(highlighted_image, cv2.COLOR_BGR2RGB))
plt.title('Highlighted Image')

plt.subplot(1, 3, 3)
plt.imshow(mask, cmap='gray')
plt.title('Mask')

plt.show()
```

#Kernal Masking Ceiling

```python
file_path = 'wall.jpg'
image = cv2.imread(file_path)
```

```python
# Adjusted color thresholds (based on understanding of dirty spots)
lower_bound = np.array([40, 30, 20])
upper_bound = np.array([180, 150, 130])
mask = cv2.inRange(image, lower_bound, upper_bound)
filled_image = image.copy()
```

```python
kernel_size = 180
kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size **2)

blurred_image = cv2.filter2D(image, -1, kernel)
```

```python
# Replace masked (dirty) pixels with blurred pixels
filled_image[mask == 255] = blurred_image[mask == 255]
```

```python
 # Plot the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Filled Image')
plt.imshow(cv2.cvtColor(filled_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Mask Image')
plt.imshow(mask, cmap='gray')
plt.axis('off')

plt.show()
```
